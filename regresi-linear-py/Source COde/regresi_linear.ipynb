```python
import matplotlib.pyplot as plt
from scipy import stats

N = int(input("Masukkan banyaknya titik data: "))

x = []
y = []
Exy = 0
Ex = 0
Ey = 0
Ex2 = 0
X = []

print("Masukkan nilai x1 sampai dengan x{}".format(N))
for i in range(N):
  x.append(float(input("")))

print("\nMasukkan nilai y1 sampai dengan y{}".format(N))
for i in range(N):
  y.append(float(input("")))

for i in range(N):
  Ex = Ex+x[i]
  Ey = Ey+y[i]
  Ex2 = Ex2+x[i]*x[i]
  xy = x[i]*y[i]
  Exy = Exy+xy;

m = (N*Exy-Ex*Ey)/(N*Ex2-Ex*Ex)
c = (Ey/N)-m*(Ex/N)

print("\nSehingga persamaan kurva linier :")
print("y = {:.6f}x + {:.6f}".format(m, c))

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y, color='red')
plt.plot(x, mymodel)
plt.show()

N = int(input("\nMasukkan banyaknya titik data hasil regresi: "))

for i in range(N):
  X.append(float(input("")))

print("\nTabel Data Hasil Regresi")
print("|No\t|x\t\t|y\t\t|")
for i in range(N):
  print("|{}\t|{:.6f}\t|{:.6f}\t|".format(i+1, X[i], m*X[i]+c))
```
